```python
import numpy as np
from StringIO import StringIO 
 
pfile=open('datos_r_2[i].txt','r')
 
data=pfile.read()
 
pfile.close()
 
data=np.genfromtxt(StringIO(data)).T #Se sobre entiende que los 
                                   #delimitadores son espacios
 
print data
```
